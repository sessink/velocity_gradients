```python
def alphabet_label(fig):
    ''' Give each subplot a alphabetic label at top left corner'''
    for i,ax in enumerate( fig.get_axes() ):
        ax.annotate(chr(i+65),(0,1.02),xycoords='axes fraction',weight='bold')
```
